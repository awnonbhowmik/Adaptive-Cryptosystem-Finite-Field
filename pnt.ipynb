```python
from matplotlib import pyplot as plt
from sympy.ntheory.primetest import isprime
from sympy.ntheory.generate import primepi
import numpy as np
```

```python
p=[]
upper_limit=1000

for x in range(2,upper_limit):
    if isprime(x):
        p.append(x)

t=np.linspace(2,upper_limit,1000)
y1=1/np.log(t)
y2=[primepi(t)/t for t in range(upper_limit)]

# print(len(t),len(y1),len(y2))


fig,axis=plt.subplots(1,2,figsize=(15,5))

plt.figure()
plt.subplot(1,2,1)
axis[0].set_title('A')
axis[0].plot(t, y1)
axis[0].plot(t, y2)
axis[0].legend(['li(x)','$\dfrac{\pi(x)}{x}$'],loc='best')

y3=[primepi(t) for t in range(upper_limit)]
y4=t/np.log(t)

plt.subplot(1,2,2)
axis[1].set_title('B')
axis[1].plot(t, y3)
axis[1].plot(t, y4)
axis[1].legend(['$\pi(x)$','$\dfrac{x}{\ln(x)}$'],loc='best')

plt.setp(axis[0],xlabel='Search Size $x$')
plt.setp(axis[0],ylabel='Prime Density')

plt.setp(axis[1],xlabel='Search Size $x$')
plt.setp(axis[1],ylabel='Number of Primes $<x$')

fig.savefig('Fig1.png',dpi=190)
```
