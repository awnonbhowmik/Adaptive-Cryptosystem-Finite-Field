```python
from matplotlib import pyplot as plt
from sympy.ntheory.primetest import isprime
from sympy.ntheory.generate import primepi
import numpy as np
```

```python
p=[]
upper_limit=1000

for x in range(2,upper_limit):
    if isprime(x):
        p.append(x)

t=np.linspace(2,upper_limit,1000)
y1=1/np.log(t)
y2=[primepi(t)/t for t in range(upper_limit)]

# print(len(t),len(y1),len(y2))

plt.figure(dpi=900)
plt.title('$\dfrac{\pi(x)}{x}\sim$li(x)')
plt.plot(t, y1)
plt.plot(t, y2)
#plt.fill_between(t,y1,y2,color='green',alpha=0.5)
plt.legend(['li(x)','$\dfrac{\pi(x)}{x}$'],loc='best')
plt.xlabel('x')
plt.ylabel('Prime Density')
plt.savefig('Fig1a.png')

y3=[primepi(t) for t in range(upper_limit)]
y4=t/np.log(t)

plt.figure(dpi=900)
plt.title('$\pi(x)\sim\dfrac{x}{\ln(x)}$')
plt.plot(t, y3)
plt.plot(t, y4)
#plt.fill_between(t,y3,y4,where=[(y3<=y4) for t in t],color='green',alpha=0.25)
plt.legend(['$\pi(x)$','$\dfrac{x}{\ln(x)}$'],loc='best')
plt.xlabel('x')
plt.ylabel('Number of primes < x')
plt.savefig('Fig1b.png')
```
